**Описание проекта:**

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно **контролировать температуру сплава**. Ваша задача — построить модель, которая будет её предсказывать **(целевой признак - температура сплава)**. 
Заказчик хочет использовать разработанную модель для имитации технологического процесса.

**Описание этапа обработки:**

Сталь обрабатывают **в металлическом ковше вместимостью около 100 тонн**. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и **подогревают до нужной температуры графитовыми электродами**. Они установлены в крышке ковша. 
Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).
Перед тем как первый раз ввести легирующие добавки, **измеряют температуру стали** и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и **снова проводят измерения**. Такой **цикл повторяется** до достижения целевого химического состава и оптимальной температуры плавки.
Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

### Данные об электродах

Немного теории.

В сети переменного тока на полезную работу затрачивается не вся, а только часть мощности (это активная мощность).

**Активная (P)** мощность - это физическая величина, характеризующая скорость генерации тока или количество переданной, либо потреблённой электроэнергии в единицу времени.

При наличии в электрической цепи переменного тока ёмкостных или индуктивных нагрузок, появляются паразитные токи, не участвующие в выполнении полезной работы. Мощность этих токов называют **реактивной (Q)**.

**Полная мощность (S):**
$$
S = \sqrt{P^2 + Q^2}
$$
, ее предприятие оплачивает.

```python
df_info(df_1)
```

Пропусков и полных дубликатов нет. Минимальное значение реактивной мощности меньше нуля, чего в нашем случае не может быть - выберем такие ошибки и исправим их:

```python
df_1.loc[df_1['реактивная_мощность'] < 0]
```

```python
k = (df_1.loc[df_1.index != 9780, 'реактивная_мощность'] / df_1.loc[df_1.index != 9780, 'активная_мощность']).median()
print(f'Медианное значение соотношения реактивной и активной мощностей равно: {k:.2f}')
```

Для дальнейшего анализа и прогнозирования важны не даты начала и конца нагрева дугой, а продолжительность нагрева. Соответственно, добавим колонку с расчетом продолжительности нагрева в секундах и удалим две колонки с датами и временем. Также, рассчитаем колонку с полной мощностью и удалим колонки с активной и реактивной мощностями:  

```python
df_1_group = df_1.copy()
df_1_group['продолжительность_нагрева'] = (
    df_1_group['конец_нагрева_дугой'] - df_1_group['начало_нагрева_дугой']).apply(
    lambda x: x.total_seconds()).astype(int)
df_1_group['полная_мощность'] = (
    df_1_group['активная_мощность']**2 + df_1_group['реактивная_мощность']**2)**0.5
df_1_group = df_1_group[['key', 'продолжительность_нагрева', 'полная_мощность']]
df_1_group.head(3)
```

```python
# функция построения гистограммы распределения и диаграммы размаха по данным столбца

def hist_box_plot(column: pd.Series, indicator: str):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15, 5))

    ax1.set_title('Гистограмма распределения', weight='bold')
    ax1.set_xlabel(indicator)
    ax1.set_ylabel('Частота')
    ax1.grid(True)
    sns.histplot(data=column, kde=True, ax=ax1)

    ax2.set_title('Диаграмма размаха', weight='bold')
    sns.boxplot(data=column, ax=ax2)
    ax2.set(xticklabels=[])
    ax2.grid(True)
    ax2.set_ylabel(indicator)

    plt.show()
```

```python
# проверки временного ряда на хронологический порядок и повторы
def check_datetime(column):
    yes_no = lambda x: 'да' if x else 'нет'
    print('Даты и время расположены в хронологическом порядке:',  yes_no(column.is_monotonic_increasing))
    print('Даты и время не повторяются:',  yes_no(column.is_unique))
```

```python
# сортируем дату и время в хронологическом порядке
df_5 = df_5.sort_values(by='время_замера')

# создаем Series с начальными и конечными временными данными
df_5_time_start = df_5.groupby(['key'])['время_замера'].first()
df_5_time_finish = df_5.groupby(['key'])['время_замера'].last()
df_1_time_start = df_1.groupby(['key'])['начало_нагрева_дугой'].first()
df_1_time_finish = df_1.groupby(['key'])['конец_нагрева_дугой'].last()

# создаем датафрейм для дальнейшего добавления начальной и конечной температур
df_5_group = pd.DataFrame(index=df_5_time_start.index, columns=['температура_старт', 'температура_финиш'])

# проверяем, что дата и время старта/финиша меньше/больше даты и времени начала/конца нагрева дугой
# если в df_1 нет данных по ковшу - оставляем NaN
# заполняем столбцы с температурой старта и финиша

for key in df_5_group.index:
    if key in df_1_time_start.index:
        if df_5_time_start[key] < df_1_time_start[key]:
            df_5_group.loc[key, 'температура_старт'] = (
                float(df_5.loc[df_5['время_замера'] == df_5_time_start[key]]['температура'])
            )
        if df_5_time_finish[key] > df_1_time_finish[key]:
            df_5_group.loc[key, 'температура_финиш'] = (
                float(df_5.loc[df_5['время_замера'] == df_5_time_finish[key]]['температура'])
            )

# удаляем иерархию в названии колонок
df_5_group = df_5_group.reset_index()

df_5_group
```

## Обучение моделей машинного обучения

### Подготовка данных для обучения моделей

```python
features = df.drop(['температура_финиш'], axis=1)
target = df['температура_финиш']

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE)

print(f'Исходный датасет: {df.shape}')
print(f'Проверка: features_train {features_train.shape}, target_train {target_train.shape}')
print(f'Проверка: features_test {features_test.shape}, target_test {target_test.shape}')
```

**Выводы:**
- данные разделены на тренировочную и тестовую выборки в соотношении 3:1

### Модель LinearRegression

```python
%%time

pipe_lr = make_pipeline(StandardScaler(), LinearRegression())

cvs_lr = cross_val_score(pipe_lr, features_train, target_train, scoring = 'neg_mean_absolute_error', n_jobs = -1)

model_lr_mae = abs(cvs_lr).mean()

print(f'Лучшее значение MAE: {model_lr_mae:0.2f}')
print('*' * 50)
```

### Модель RandomForestRegressor

```python
# определяем целевую функцию
def objective_rfr(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 250)
    max_depth = trial.suggest_int('max_depth', 1, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)

    pipe_rfr = Pipeline(
        [('StandardScaler', StandardScaler()), 
         ('RandomForestRegressor', RandomForestRegressor(
             n_estimators=n_estimators,
             max_depth=max_depth, 
             min_samples_leaf=min_samples_leaf, 
             random_state=RANDOM_STATE, 
             verbose=False))
        ]
    )
    
    score = cross_val_score(
        pipe_rfr, 
        features_train, 
        target_train, 
        scoring = 'neg_mean_absolute_error',
        error_score='raise',
        n_jobs = -1).mean()
    
    return score
```

```python
%%time

# создаем объект исследования
study_rfr = optuna.create_study(study_name="rfr_optimization",
                                direction="maximize" #, sampler=TPESampler()
                               )

# запускаем процесс оптимизации
study_rfr.optimize(objective_rfr, n_trials=20, show_progress_bar=True)
```

```python
study_rfr.trials_dataframe()
```

```python
print(f'Лучшее значение MAE: {-study_rfr.best_value:.2f}')
print(f'Гиперпараметры модели: {study_rfr.best_trial.params}')
```

### Модель CatBoostRegressor

```python
# определяем целевую функцию
def objective_cbr(trial):
    iterations = trial.suggest_int('iterations', 180, 220)
    depth = trial.suggest_int('depth', 2, 10)
    learning_rate = trial.suggest_float('learning_rate', 0.05, 0.1, step=0.01)

    pipe_cbr = Pipeline(
        [('StandardScaler', StandardScaler()), 
         ('CatBoostRegressor', CatBoostRegressor(
             iterations=iterations,
             depth=depth, 
             learning_rate=learning_rate, 
             random_state=RANDOM_STATE, 
             verbose=False))
        ]
    )
    
    score = cross_val_score(
        pipe_cbr, 
        features_train, 
        target_train, 
        scoring = 'neg_mean_absolute_error',
        error_score='raise',
        n_jobs = -1).mean()
    
    return score
```

```python
%%time

# создаем объект исследования
study_cbr = optuna.create_study(study_name="cbr_optimization",
                                direction="maximize"
                               )

# запускаем процесс оптимизации
study_cbr.optimize(objective_cbr, n_trials=10, show_progress_bar=True)
```

```python
study_cbr.trials_dataframe()
```

```python
print(f'Лучшее значение MAE: {-study_cbr.best_value:.2f}')
print(f'Гиперпараметры модели: {study_cbr.best_trial.params}')
```
