Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком.

### Изучение моделей без учёта дисбаланса

```python
# функция для расчета основных параметров моделей
def culc_values(target_valid, predicted_valid, model): 
    print(model)
    print('*' * 50)
    print('Доля правильных ответов - Accuracy: {:.3f}'.format(accuracy_score(target_valid, predicted_valid)))
    print('*' * 50)
    print('Матрица ошибок - [[TN FN] [FP TP]]: \n', confusion_matrix(target_valid, predicted_valid))
    print('*' * 50)
    print('Полнота - Recall = TP/(TP + FN): {:.3f}'.format(recall_score(target_valid, predicted_valid)))
    print('Точность - Precision = TP/(TP + FP):{:.3f}'.format(precision_score(target_valid, predicted_valid)))
    print('*' * 50)
    print('F1-мера - F1 = 2 * Precision * Recall/(Precision + Recall):{:.3f}'.format(
        f1_score(target_valid, predicted_valid)))
    print('*' * 50)
```

```python
def hyperpar_selection_DTC(
    features_train, target_train, features_valid, target_valid, rnd_state=12345, cl_weight=None):
    best_model_DTC = None
    best_result_DTC = 0
    for depth in range(1, 14):
        for crit in ['gini']: #'entropy'
            for msl in range(1, 11):
                model = DecisionTreeClassifier(random_state=rnd_state, criterion=crit, max_depth=depth, \
                                               min_samples_leaf = msl, class_weight=cl_weight)
                model.fit(features_train, target_train)
                predicted_valid = model.predict(features_valid)
                result = accuracy_score(target_valid, predicted_valid)
                if result > best_result_DTC:
                    best_model_DTC = model
                    best_result_DTC = result
    return best_model_DTC, predicted_valid
```

```python
%%time

best_model_DTC_0, predicted_valid = hyperpar_selection_DTC(
    features_train, target_train, features_valid, target_valid)

culc_values(target_valid, predicted_valid, best_model_DTC_0)
```

```python
def hyperpar_selection_RFC(
    features_train, target_train, features_valid, target_valid, rnd_state=12345, cl_weight=None):
    best_model_RFC = None
    best_result_RFC = 0
    for est in [50, 100]:
        for depth in range(1, 14):
            for crit in ['gini']: #'entropy'
                for msl in range(1, 11):
                    model = RandomForestClassifier(random_state=rnd_state, n_estimators=est, criterion=crit, \
                                                   max_depth=depth, min_samples_leaf = msl, class_weight=cl_weight)
                    model.fit(features_train, target_train)
                    predicted_valid = model.predict(features_valid)
                    result = accuracy_score(target_valid, predicted_valid)
                    if result > best_result_RFC:
                        best_model_RFC = model
                        best_result_RFC = result
    return best_model_RFC, predicted_valid
```

```python
%%time

best_model_RFC_0, predicted_valid = hyperpar_selection_RFC(
    features_train, target_train, features_valid, target_valid)

culc_values(target_valid, predicted_valid, best_model_RFC_0)
```

```python
def hyperpar_selection_LR(
    features_train, target_train, features_valid, target_valid, rnd_state=12345, cl_weight=None):
    best_model_LR = None
    best_result_LR = 0
    for m_iter in [60, 100, 200]:
        model = LogisticRegression(random_state=rnd_state, max_iter=m_iter, class_weight=cl_weight)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        result = accuracy_score(target_valid, predicted_valid)
        if result > best_result_LR:
            best_model_LR = model
            best_result_LR = result
    return best_model_LR, predicted_valid
```

```python
%%time

best_model_LR_0, predicted_valid = hyperpar_selection_LR(
    features_train, target_train, features_valid, target_valid)

culc_values(target_valid, predicted_valid, best_model_LR_0)
```

### Увеличение выборки

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

### Уменьшение выборки

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```
